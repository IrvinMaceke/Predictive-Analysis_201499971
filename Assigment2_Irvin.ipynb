```python
from statsmodels.tsa.stattools import adfuller


result = adfuller(daily_steps['Actual'])


adf_statistic = result[0]
p_value = result[1]
critical_values = result[4]


print(f'ADF Statistic: {adf_statistic}')
print(f'P-value: {p_value}')
print('Critical Values:')
for key, value in critical_values.items():
    print(f'   {key}: {value}')
    

if p_value <= 0.05:
    print("Reject null hypothesis: Data is stationary")
else:
    print("Fail to reject null hypothesis: Data is not stationary")
```

```python
# Estimating Model Parameters (ARIMA)

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plot_acf(daily_steps['Actual'], lags=30, ax=plt.gca(), title=' autocorrelation function(ACF)')

plt.subplot(2, 1, 2)
plot_pacf(daily_steps['Actual'], lags=30, ax=plt.gca(), title='partial autocorrelation function (PACF)')

plt.tight_layout()
plt.show()
```

```python


# Estimating the model parameters for an ARIMA model, and it includes generating the autocorrelation function (ACF) and partial 
# autocorrelation function (PACF) plots. These plots are important for understanding the underlying patterns and relationships 
# in the time series data and help in determining appropriate values for the ARIMA model's parameters. 




```

```python


from statsmodels.tsa.arima.model import ARIMA

p = 1  
d = 1  
q = 1  

model = ARIMA(daily_steps['Actual'], order=(p, d, q))
model_fit = model.fit()

print(model_fit.summary())



plt.figure(figsize=(12, 8))


plt.subplot(2, 1, 1)
plt.plot(daily_steps['Date'], daily_steps['Actual'], label='Actual Steps', color='blue')
plt.title('Growth of Actual Steps Over Time')
plt.xlabel('Date')
plt.ylabel('Steps')
plt.legend()


plt.subplot(2, 1, 2)
plt.plot(daily_steps['Date'], model_fit.fittedvalues, label='Fitted ARIMA Model', color='orange')
plt.title('Fitted ARIMA Model')
plt.xlabel('Date')
plt.ylabel('Steps')
plt.legend()

plt.tight_layout()
plt.show()



```

```python



# The results presented reflect the outcome of applying a SARIMAX model to the time series data. The model parameters (1, 1, 1) 
# denote the autoregressive, differencing, and moving average orders respectively. The log likelihood, AIC, 
# and BIC are metrics assessing how well the model fits the data, with lower values indicating a better fit. Coefficients 
# represent the influence of autoregressive and moving average terms. Sigma squared quantifies residual variance. 
# The Ljung-Box test checks for residual autocorrelation, while the Jarque-Bera test assesses normality of residuals. 
# Heteroskedasticity tests for varying residual variance. Overall, these results provide insights into model 
# quality, coefficient significance, and potential violations of underlying assumptions.



```

```python

import pmdarima as pm
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt



column_name = 'Actual'


ts_data = combined_data[column_name]


train_size = int(0.8 * len(ts_data))
train_data, test_data = ts_data.iloc[:train_size], ts_data.iloc[train_size:]


auto_sarima_model = pm.auto_arima(train_data, seasonal=True, m=12, stepwise=True, suppress_warnings=True)


print(auto_sarima_model.summary())


forecast_steps = len(test_data)
forecast, conf_int = auto_sarima_model.predict(n_periods=forecast_steps, return_conf_int=True)


rmse = np.sqrt(mean_squared_error(test_data, forecast))
print("Root Mean Squared Error (RMSE):", rmse)


plt.figure(figsize=(10, 6))
plt.plot(test_data.index, test_data, label='Actual Steps', color='green')
plt.plot(test_data.index, forecast, label='Forecasted Steps', linestyle='--', color='blue')
plt.fill_between(test_data.index, conf_int[:, 0], conf_int[:, 1], color='blue', alpha=0.2)
plt.title('Actual vs. Forecasted Steps')
plt.xlabel('Date')
plt.ylabel('Steps')
plt.legend()
plt.grid()
plt.show()




```
